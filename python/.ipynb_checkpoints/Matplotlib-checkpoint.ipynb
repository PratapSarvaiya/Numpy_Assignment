```python
x = [1,2,3,4,5]
y1 = [5,1,7,8,2]
y2 = [7,8,3,5,1]

plt.plot(x,y1, color = "red",marker = "o", linestyle = "--", linewidth = 3, markersize = 6, label = "liney1")
```

```python
x = [1,2,3,4,5]
y1 = [5,1,7,8,2]
y2 = [7,8,3,5,1]

plt.plot(x,y1, color = "red",marker = "o", linestyle = "--", linewidth = 3, markersize = 6, label = "line y1 as tata motors")
plt.plot(x,y2, color = "b",marker = "o", linestyle = "--", linewidth = 3, markersize = 6, label = "line y2 as tata power")

plt.title("Multiple lines in one plot")

plt.xlabel("Days")
plt.ylabel("stock price")
plt.legend()
plt.show()
```

```python
plt.figure()  # initialising a page to make plot
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))

plt.show()
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))

plt.subplot(2,2,3)
plt.plot(x,np.tan(x))

plt.subplot(2,2,4)
plt.plot(x,np.sin(x+1))

plt.show()
```

```python
# bar plot >> to see the distribution of categorical variables

x = ["a","b","c","d"]  # name of students
y = [4, 6, 1, 10]  # no of days present in the class

plt.bar(x, y, color = 'r')  # vertical plot
```

```python
# pie chart >> to understand the percent distribution of data

langs = ["c", 'python', "java"]
students = [20, 100, 40]

plt.pie(students, labels = langs)
plt.title("pie chart")
plt.show()
```

```python
langs = ["c", 'python', "java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%")  # percentage in view
plt.title("pie chart")
plt.show()
```

```python
# focus on one segment

explode = (0.0, 0.1,0.1)
langs = ["c", 'python', "java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%, explode = explode")  # percentage in view
plt.title("pie chart")
plt.show()
```

```python
explode = (0.0, 0.1,0.1)
langs = ["c", 'python', "java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%, explode = explode, shadow = True")  # percentage in view
plt.title("pie chart")
plt.show()
```

```python
# 3d plot >> 3 dimension plot
x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x, y, z, c='r')
```

```python
years = [2021,2016,2022,2023,2017]
category1 = [20, 25, 30, 28, 35]
category2 = [10, 15, 20, 18, 25]
category3 = [5, 8, 10, 9, 15]

plt.stackplot(years, category1,category2,category3,labels = ["category1","category2","category3",],
             colors = ["skyblue","lightgreen","orange"])
plt.xlabel("years")
plt.ylabel("Values")
plt.title("Area plot")
plt.legend()
plt.show()
```
