```python
#how tip is varying with respect to total_bill >> scatterplot by default relplot makes scatter plot
sns.relplot(x='total_bill', y = 'tip',data = df)  # relationship between two variables, by default it gives scatter plot, kind='line' 
# insight >> as total bill increases tip increases
```

```python
sns.relplot(x='total_bill', y = 'tip',data = df,hue='smoker') # hue gives the different color shades
#insight >>  customer who smoked gave the highest tip
```

```python
sns.relplot(x='total_bill', y = 'tip',data = df, hue = 'size') # different color coding column value
```

```python
sns.relplot(x='total_bill', y = 'tip',data = df, style='size')  # style gives the different representation of columns value
```

```python
sns.relplot(x='total_bill', y = 'tip',data = df, style='smoker',hue='time') 
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# two subplot based on a columns
sns.relplot(x='total_bill',y='tip',hue='smoker',data=tips, col = 'size')
```

```python
# categorical data plots
```

```python
tips
```

```python
tips.dtypes
```

```python
# for categorical column >> cat plot
# plot between data and total bill, day is a categorical columns
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
# insights >> many people arrives on weekends as compared to weekdays
# in cat plot, few datapoints are overlapping>> to see it separately use swarm plot

sns.catplot(x='day',y='total_bill',data=tips, kind='swarm')
```

```python
# boxplot >> to see the outliers
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
# boxen plot is made from box plot
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen')
```

```python
tips
```

```python
# how total bill(aggregated) varies with different data
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
# draw a plot of two variables with distribution >>bivariant plot
sns.jointplot(x=tips.total_bill, y=tips.tip) 
```

```python
# draw a plot of two variables with distribution >>bivariant plot means two plot
sns.jointplot(x=tips.total_bill, y=tips.tip) 
```

```python
# joints plot >> understand relationship and also distribution of two variables

sns.jointplot(x=tips.total_bill, y=tips.tip, kind = 'reg')  # reg >> regression line represent relationship
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip, kind = 'hex')
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip, kind = 'kde') # kernel density estimeter, contour >> metrological department
```
