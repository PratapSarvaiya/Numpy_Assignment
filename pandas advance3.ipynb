```python
inner merge >> merge based on common column is called inner merge(intersection)
left merge >> All the left elements of dataframe  A will be in the merge
right merge >> All the element of right dataframe B will be in the merge
Outer merge >> All the element of both the dataframe
cross >> every permutation of rows are there
```

```python
# merge

df1 = pd.DataFrame({'key1':[1,2,4,5,6],
                     'key2':[4,5,6,7,8],
                     'key3':[3,4,5,6,6]})
df1
```

```python
df2 = pd.DataFrame({'key1':[1,2,45,6,67],
                     'key4':[56,5,6,7,8],
                     'key5':[3,56,5,6,6]})
df2
```

```python
pd.merge(df1,df2,how='inner')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer') # all the common element of the common column from both the dataframe
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how='cross')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how='left',left_on="key2",right_on="key4")
```
